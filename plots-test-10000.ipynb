```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Provided data
probabilities_with_holo = [9.61, 9.37, 12.29, 9.01, 8.78, 5.60, 5.86, 7.79, 2.76, 11.34, 7.91, 8.39, 9.13, 9.56, 5.53]
probabilities_without_holo = [2.69, 0.45, 3.98, 0.41, 0.96, 3.20, 5.76, 2.46, 3.73, 0.89, 1.70, 0.8, 4.83, 21.27, 3.28]

# Combining the data into a DataFrame for easier plotting with Seaborn
data = pd.DataFrame({
    'Probabilities': probabilities_with_holo + probabilities_without_holo,
    'Category': ['With Thing']*15 + ['Without Thing']*15
})

# Threshold value
threshold = 4.75

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Probabilities', data=data)
plt.axhline(y=threshold, color='r', linestyle='--')
plt.title('Box Plot of Probabilities')
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(probabilities_with_holo, color='blue', label='With Thing', kde=True)
sns.histplot(probabilities_without_holo, color='orange', label='Without Thing', kde=True)
plt.axvline(x=threshold, color='r', linestyle='--')
plt.legend()
plt.title('Histogram of Probabilities')
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(range(len(probabilities_with_holo)), probabilities_with_holo, color='blue', label='Video with hologram')
plt.scatter(range(len(probabilities_without_holo)), probabilities_without_holo, color='orange', label='Video without hologram')
#plt.axhline(y=threshold, color='r', linestyle='--')
plt.legend()
plt.xlabel('Video index')
plt.ylabel('Hologram probability %')
plt.title('Scatter Plot of Hologram Presence Probabilities per Video')
plt.show()

# Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(probabilities_with_holo, color='blue', label='With Thing', shade=True)
sns.kdeplot(probabilities_without_holo, color='orange', label='Without Thing', shade=True)
plt.axvline(x=threshold, color='r', linestyle='--')
plt.legend()
plt.title('Density Plot of Probabilities')
plt.show()
```
